```python
from tkinter import *
from PIL import Image, ImageTk
import cv2
import numpy as np

H, S, V = 128, 128, 128


class Window(Frame):
    
    def __init__(self, master = None):
        Frame.__init__(self, master)
        self.master = master   
        self.init_window()
    
    def init_window(self):
        self.master.title('GUI')
        self.pack(fill=BOTH, expand=1)
        
        leftButton1 = Button(self, text='Less H', command=self.lessH)
        leftButton1.place(x=240, y=500)
        rightButton1 = Button(self, text='More H', command=self.moreH)
        rightButton1.place(x=450, y=500)
        leftButton2 = Button(self, text='Less S', command=self.lessS)
        leftButton2.place(x=240, y=550)
        rightButton2 = Button(self, text='More S', command=self.moreS)
        rightButton2.place(x=450, y=550)
        leftButton3 = Button(self, text='Less V', command=self.lessV)
        leftButton3.place(x=240, y=600)
        rightButton3 = Button(self, text='More V', command=self.moreV)
        rightButton3.place(x=450, y=600)
        
        plot = Button(self, text='Image', command=self.plotImage)
        plot.place(x=450, y=600)
        
        menu = Menu(self.master)
        self.master.config(menu=menu)
        
        file = Menu(menu)
        file.add_command(label="Save", command=self.client_exit)
        file.add_command(label="Exit", command=self.destroy)
        menu.add_cascade(label='File', menu=file)
        
        edit = Menu(menu)
        edit.add_command(label='Show Image', command=self.showImg())
        edit.add_command(label='Show Text', command=self.showTxt())
        menu.add_cascade(label='Edit', menu=edit)
        
    def client_exit(self):
        Frame.destroy()
        
    def showImg(self):
        load = Image.open('containers1.jpg').resize((400,400))
        #load_cvt = Image.c
        #load = np.array(load)
        #load = cv2.cvtColor(load, cv2.COLOR_RGB2HSV)
        render = ImageTk.PhotoImage(load)
        img = Label(self, image=render)
        img.image = render
        img.place(x=175, y=50)
        
    def plotImage(self):
        load = Image.open('containers1.jpg').resize((400,400))
        #load = np.array(load)
        #load = cv2.cvtColor(load, cv2.COLOR_RGB2HSV)
        render = ImageTk.PhotoImage(load)
        img = Label(self, image=render)
        img.image = render
        img.place(x=175, y=50)
        
    def showTxt(self):
        text = Label(self, text='HSV Colorspace!')
        text.pack()
        
    def lessH(self):
        text = Label(self)
        text.pack()
        text.place(x=400,y=500)
        dec_label(text)
        
        
    def moreH(self):
        text = Label(self)
        text.pack()
        text.place(x=400,y=500)
        inc_label(text)
        
    def lessS(self):
        text = Label(self)
        text.pack()
        text.place(x=400,y=550)
        dec_label(text)
        
        
    def moreS(self):
        text = Label(self)
        text.pack()
        text.place(x=400,y=550)
        inc_label(text)
        
    def lessV(self):
        text = Label(self)
        text.pack()
        text.place(x=400,y=600)
        dec_label(text)
        
        
    def moreV(self):
        text = Label(self)
        text.pack()
        text.place(x=400,y=600)
        inc_label(text)
        
def dec_label_H(label):
    def count():
        global H
        H = max(0, H-1)
        label.config(text=str(H))
        label.after(func=count)
    count()
    
def inc_label_H(label):
    def count():
        global H
        H = min(255, H+1)
        label.config(text=str(H))
        label.after(func=count)
    count()

def dec_label_S(label):
    def count():
        global S
        S = max(0, S-1)
        label.config(text=str(S))
        label.after(func=count)
    count()
    
def inc_label_S(label):
    def count():
        global S
        S = min(255, S+1)
        label.config(text=str(S))
        label.after(func=count)
    count()
    
def dec_label_V(label):
    def count():
        global V
        V = max(0, V-1)
        label.config(text=str(V))
        label.after(func=count)
    count()
    
def inc_label_V(label):
    def count():
        global V
        V = min(255, V+1)
        label.config(text=str(V))
        label.after(func=count)
    count()

root = Tk()
root.geometry("750x700")
app = Window(root)
root.mainloop()
```

```python
import tkinter as tk

counter = 0 
def counter_label(label):
  def count():
    global counter
    counter += 1
    label.config(text=str(counter))
    #label.after(1000, count)
  count()
 
 
root = tk.Tk()
root.title("Counting Seconds")
label = tk.Label(root, fg="green")
label.pack()
counter_label(label)
button = tk.Button(root, text='Stop', width=25, command=root.destroy)
button.pack()
root.mainloop()
```
